```python
#Looks up page corresponding to an Amazon search of printers
def get_soup_iprinter(page = None):
    #Sets Up Query For Page Number
    #searh for printer in Inkjet Computer Printers, New Ink Jet Printers, 3 Stars and up, sorted by average customer reviews
    if page == None or page < 2:
        r = requests.get('https://www.amazon.com/s?k=printer&i=office-electronics&rh=n%3A1064954%2Cn%3A172646&s=review-rank&dc&qid=1555354437&rnid=2941120011&ref=sr_st_review-rank')
        c = r.content
        return BeautifulSoup(c, 'html.parser')
    
    #Check if page is > 2
    else:
        r = requests.get('https://www.amazon.com/s?k=printer&i=office-electronics&rh=n%3A1064954%2Cn%3A172646&s=review-rank&dc&page={}&qid=1555354452&rnid=2941120011&ref=sr_pg_{}'.format(page, page))
        c = r.content
        return BeautifulSoup(c, 'html.parser')
```
